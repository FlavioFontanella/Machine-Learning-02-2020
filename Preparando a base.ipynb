```python
mgmc19 = []
mgm38 = []
mgsc19 = []
mgs38 = []

mvc19 = []
mv38 = []
mec19 = []
me38 = []
mdc19 = []
md38 = []


for i in range(len(base)-380):
    ano, rodada, mandante = base.loc[i+380].ano, base.loc[i+380]["round"], base.loc[i+380].home_team
    bas19 = base[base["ano"]<ano][base["home_team"]==mandante].append(base[base["ano"]==ano][base["round"]<rodada][base["home_team"]==mandante])[-19:]
    bas38 = base[base["ano"]<ano][base["home_team"]==mandante].append(base[base["ano"]==ano][base["round"]<rodada][base["home_team"]==mandante]).append(base[base["ano"]<ano][base["away_team"]==mandante]).append(base[base["ano"]==ano][base["round"]<rodada][base["away_team"]==mandante]).sort_index()[-38:]
    gm19 = bas19["placar_casa"].sum()
    gm38 = bas38[bas38["home_team"]==mandante]["placar_casa"].sum() + bas38[bas38["away_team"]==mandante]["placar_visitante"].sum()
    if gm19>19:
        mgmc19.append(gm19)
    else:
        mgmc19.append(19)
    if gm38>30:
        mgm38.append(gm38)
    else:
        mgm38.append(30)
    gs19 = bas19["placar_visitante"].sum()
    gs38 = bas38[bas38["home_team"]==mandante]["placar_visitante"].sum() + bas38[bas38["away_team"]==mandante]["placar_casa"].sum()
    if gs19>9:
        mgsc19.append(gs19)
    else:
        mgsc19.append(27)
    if gs38>24:
        mgs38.append(gs38)
    else:
        mgs38.append(64)
    v19 = bas19[bas19["resultado"]==1]["resultado"].sum()
    e19 = bas19[bas19["resultado"]==2]["resultado"].sum()/2
    d19 = bas19[bas19["resultado"]==3]["resultado"].sum()/3
    mvc19.append(v19)
    mec19.append(e19)
    mdc19.append(d19)
    v38 = bas38[bas38["home_team"]==mandante][bas38["resultado"]==1]["resultado"].sum() + bas38[bas38["away_team"]==mandante][bas38["resultado"]==3]["resultado"].sum()/3
    e38 = bas38[bas38["resultado"]==2]["resultado"].sum()/2
    d38 = bas38[bas38["home_team"]==mandante][bas38["resultado"]==3]["resultado"].sum()/3 + bas38[bas38["away_team"]==mandante][bas38["resultado"]==1]["resultado"].sum()
    mv38.append(v38)
    me38.append(e38)
    md38.append(d38)
            
```
